```python
import folium
from folium.plugins import MarkerCluster

# Centrar mapa
lat_centro = df['LATITUD_NUM'].mean()
lon_centro = df['LONGITUD_NUM'].mean()
mapa = folium.Map(location=[lat_centro, lon_centro], zoom_start=6)

cluster = MarkerCluster().add_to(mapa)

# Función para colorear según ratio
def color_por_ratio(ratio):
    if ratio < 0.9:
        return 'red'
    elif ratio < 1.1:
        return 'orange'
    elif ratio < 2.0:
        return 'green'
    else:
        return 'blue'

# Agregar marcadores
for _, row in df.iterrows():
    color = color_por_ratio(row['RATIO_CUMPLIMIENTO'])

    popup_info = f"""
    <b>Tienda ID:</b> {row['TIENDA_ID']}<br>
    <b>Venta:</b> ${row['VENTA_TOTAL']:,.0f}<br>
    <b>Meta:</b> ${row['META_VENTA']:,.0f}<br>
    <b>Ratio cumplimiento:</b> {row['RATIO_CUMPLIMIENTO']:.2f}<br><br>
    <b>Nivel socioeconómico:</b> {row['NIVELSOCIOECONOMICO_DES']}<br>
    <b>Entorno:</b> {row['ENTORNO_DES']}<br>
    <b>Segmento:</b> {row['SEGMENTO_MAESTRO_DESC']}<br>
    <b>Ubicación tienda:</b> {row['LID_UBICACION_TIENDA']}
    """

    folium.CircleMarker(
        location=[row['LATITUD_NUM'], row['LONGITUD_NUM']],
        radius=6,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.8,
        popup=folium.Popup(popup_info, max_width=300)
    ).add_to(cluster)

# Mostrar mapa
mapa
```

## Codificar variables categoricas

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Columnas categóricas nominales (sin orden natural)
columnas_categoricas = [
    'ENTORNO_DES',
    'SEGMENTO_MAESTRO_DESC',
    'LID_UBICACION_TIENDA',
    'NIVELSOCIOECONOMICO_DES'
]

# Columnas numéricas features
columnas_numericas = [
    'MTS2VENTAS_NUM',
    'PUERTASREFRIG_NUM',
    'LATITUD_NUM',
    'LONGITUD_NUM',
    'META_VENTA'
]
```

### Pipeline preprocesamiento

```python
preprocesador = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), columnas_categoricas)
    ],
    remainder='passthrough'  # deja pasar las columnas numéricas sin cambio
)

```

### Muestro los nombres de las columnas nuevas codificadas

```python
# Entrenar el encoder solo con las columnas categóricas
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoder.fit(df[columnas_categoricas])

# Ver nombres de las columnas codificadas
encoded_feature_names = encoder.get_feature_names_out(columnas_categoricas)

# Mostrarlas
for name in encoded_feature_names:
    print(name)
```
