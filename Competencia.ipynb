```python
import pandas as pd
import osmnx as ox
from shapely.geometry import Point
from shapely.ops import nearest_points
from tqdm import tqdm
import geopandas as gpd


df_oxxo = pd.read_csv("DIM_TIENDA.csv")
df_oxxo = df_oxxo.dropna(subset=["LATITUD_NUM", "LONGITUD_NUM"])
df_oxxo = df_oxxo[(df_oxxo["LATITUD_NUM"] != 0) & (df_oxxo["LONGITUD_NUM"] != 0)]

# Crear GeoDataFrame
gdf_oxxo = gpd.GeoDataFrame(
    df_oxxo.copy(),
    geometry=gpd.points_from_xy(df_oxxo["LONGITUD_NUM"], df_oxxo["LATITUD_NUM"]),
    crs="EPSG:4326"
)

# Copia para coordenadas en lat/lon
gdf_oxxo_latlon = gdf_oxxo.copy()

# Reproyectar a metros
gdf_oxxo = gdf_oxxo.to_crs(epsg=3857)

# Funciones
def buscar_competencia(lat, lon, tags, radio_m=1000):
    point = (lat, lon)
    try:
        gdf = ox.features.features_from_point(point, tags=tags, dist=radio_m)
        gdf = gdf.to_crs(epsg=3857)
        gdf = gdf[gdf.geometry.type == 'Point']
        return gdf
    except:
        return gpd.GeoDataFrame(columns=['geometry'], geometry='geometry')

def calcular_distancias_a_competencia(punto, gdf_comp):
    if gdf_comp.empty:
        return None, 0
    nearest_geom = nearest_points(punto, gdf_comp.unary_union)[1]
    distancia = punto.distance(nearest_geom)
    num_dentro_radio = gdf_comp.geometry.within(punto.buffer(1000)).sum()
    return distancia, num_dentro_radio

# Inicializar columnas
gdf_oxxo["dist_comp_directa"] = None
gdf_oxxo["num_comp_directa"] = 0
gdf_oxxo["dist_comp_indirecta"] = None
gdf_oxxo["num_comp_indirecta"] = 0

# Tags OSM
tags_directa = {"shop": "convenience"}
tags_indirecta = {"shop": ["supermarket", "kiosk", "variety_store"]}

# Proceso
for idx, row in tqdm(gdf_oxxo.iterrows(), total=len(gdf_oxxo), desc=""):
    lat = gdf_oxxo_latlon.at[idx, "LATITUD_NUM"]
    lon = gdf_oxxo_latlon.at[idx, "LONGITUD_NUM"]
    punto = row.geometry

    gdf_dir = buscar_competencia(lat, lon, tags_directa)
    dist_dir, num_dir = calcular_distancias_a_competencia(punto, gdf_dir)
    gdf_oxxo.at[idx, "dist_comp_directa"] = dist_dir
    gdf_oxxo.at[idx, "num_comp_directa"] = num_dir

    gdf_ind = buscar_competencia(lat, lon, tags_indirecta)
    dist_ind, num_ind = calcular_distancias_a_competencia(punto, gdf_ind)
    gdf_oxxo.at[idx, "dist_comp_indirecta"] = dist_ind
    gdf_oxxo.at[idx, "num_comp_indirecta"] = num_ind

# Guardar sin geometry
df_final = gdf_oxxo.drop(columns=["geometry"])
df_final.to_csv("oxxo_con_competencia.csv", index=False)
```

```python
import pandas as pd
import bambi as bmb
import arviz as az
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, r2_score

# 1. Preparación de datos
print("Preparando datos...")
variables_modelo = [
    'RATIO_CUMPLIMIENTO', 'TIENDA_ID',
    'dist_comp_directa', 'num_comp_directa',
    'dist_comp_indirecta', 'num_comp_indirecta',
    'NIVELSOCIOECONOMICO_DES', 'ENTORNO_DES',
    'MTS2VENTAS_NUM', 'PUERTASREFRIG_NUM',
    'LATITUD_NUM', 'LONGITUD_NUM',
    'SEGMENTO_MAESTRO_DESC', 'LID_UBICACION_TIENDA', 'META_VENTA'
]

df_bayes = df_modelo[variables_modelo].dropna().drop_duplicates()
print(f"Número final de tiendas: {len(df_bayes)}")  # Debe ser 940

# 2. Convertir variables categóricas
categorical_vars = ['NIVELSOCIOECONOMICO_DES', 'ENTORNO_DES',
                   'SEGMENTO_MAESTRO_DESC', 'LID_UBICACION_TIENDA']
for var in categorical_vars:
    df_bayes[var] = df_bayes[var].astype('category')

# 3. Definir y ajustar el modelo
print("\nAjustando modelo bayesiano...")
formula = """
RATIO_CUMPLIMIENTO ~
dist_comp_directa + num_comp_directa +
dist_comp_indirecta + num_comp_indirecta +
NIVELSOCIOECONOMICO_DES + ENTORNO_DES +
MTS2VENTAS_NUM + PUERTASREFRIG_NUM +
SEGMENTO_MAESTRO_DESC + LID_UBICACION_TIENDA +
META_VENTA
"""

model = bmb.Model(formula, df_bayes, family="gaussian")
results = model.fit(draws=1000, tune=1000, chains=4, random_seed=42)

# 4. Generar predicciones
print("\nGenerando predicciones...")
posterior_predictive = model.predict(results, kind="pps", inplace=False)

# 5. Promedio de predicciones por tienda
y_pred = az.extract(posterior_predictive, group="posterior_predictive")["RATIO_CUMPLIMIENTO"]
y_pred_mean = y_pred.mean('sample').values
y_pred_lower = y_pred.quantile(0.03, dim='sample').values
y_pred_upper = y_pred.quantile(0.97, dim='sample').values

# 6. Crear DataFrame de resultados
results_df = pd.DataFrame({
    'TIENDA_ID': df_bayes['TIENDA_ID'].values,
    'RATIO_REAL': df_bayes['RATIO_CUMPLIMIENTO'].values,
    'RATIO_PRED': y_pred_mean,
    'IC_INF': y_pred_lower,
    'IC_SUP': y_pred_upper
})

# 7. Verificación de dimensiones
assert len(results_df) == 940, f"Error: El DataFrame tiene {len(results_df)} filas, debería tener 940"
print("\n Dimensiones correctas - 940 tiendas:")

# 8. Evaluación del modelo
print("\nEvaluación del modelo:")
mae = mean_absolute_error(results_df['RATIO_REAL'], results_df['RATIO_PRED'])
r2 = r2_score(results_df['RATIO_REAL'], results_df['RATIO_PRED'])
cobertura = ((results_df['RATIO_REAL'] >= results_df['IC_INF']) &
             (results_df['RATIO_REAL'] <= results_df['IC_SUP'])).mean()

print(f"- MAE: {mae:.4f}")
print(f"- R²: {r2:.4f}")
print(f"- Cobertura IC 94%: {cobertura:.1%}")

# 9. Visualización de resultados
plt.figure(figsize=(12, 6))
plt.scatter(results_df['RATIO_REAL'], results_df['RATIO_PRED'], alpha=0.6)
plt.plot([results_df['RATIO_REAL'].min(), results_df['RATIO_REAL'].max()],
         [results_df['RATIO_REAL'].min(), results_df['RATIO_REAL'].max()],
         'r--')
plt.xlabel('Valor Real')
plt.ylabel('Valor Predicho (Promedio)')
plt.title('Valores Reales vs. Predicciones Promediadas')
plt.grid(True)
plt.show()

# 10. Gráfico de intervalos (50 tiendas)
plt.figure(figsize=(14, 6))
sample_tiendas = results_df.sample(50, random_state=42).sort_values('RATIO_REAL')
plt.errorbar(x=sample_tiendas['TIENDA_ID'].astype(str),
             y=sample_tiendas['RATIO_PRED'],
             yerr=[sample_tiendas['RATIO_PRED'] - sample_tiendas['IC_INF'],
                   sample_tiendas['IC_SUP'] - sample_tiendas['RATIO_PRED']],
             fmt='o', label='Predicción con IC 94%')
plt.scatter(sample_tiendas['TIENDA_ID'].astype(str),
            sample_tiendas['RATIO_REAL'],
            color='red', label='Valor Real')
plt.xticks(rotation=90)
plt.legend()
plt.title('Comparación de Predicciones Promediadas vs Valores Reales (50 tiendas muestreadas)')
plt.grid(True)
plt.tight_layout()
plt.show()

# 11. Análisis de residuos
results_df['RESIDUO'] = results_df['RATIO_REAL'] - results_df['RATIO_PRED']
plt.figure(figsize=(12, 5))
plt.scatter(results_df['RATIO_PRED'], results_df['RESIDUO'], alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Valor Predicho')
plt.ylabel('Residuo (Real - Predicho)')
plt.title('Análisis de Residuos')
plt.grid(True)
plt.show()
```

```python
import pymc as pm
import pymc_bart as pb
import numpy as np
import pandas as pd
import arviz as az
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

# 1. Preprocesamiento de datos
print("🔧 Preprocesando datos...")
variables_modelo = [
    'RATIO_CUMPLIMIENTO', 'TIENDA_ID',
    'dist_comp_directa', 'num_comp_directa',
    'dist_comp_indirecta', 'num_comp_indirecta',
    'NIVELSOCIOECONOMICO_DES', 'ENTORNO_DES',
    'MTS2VENTAS_NUM', 'PUERTASREFRIG_NUM',
    'LATITUD_NUM', 'LONGITUD_NUM',
    'SEGMENTO_MAESTRO_DESC', 'LID_UBICACION_TIENDA', 'META_VENTA'
]

# Cargar y limpiar datos
df_bart = df_modelo[variables_modelo].dropna().drop_duplicates()

# Codificar variables categóricas
cat_vars = ['NIVELSOCIOECONOMICO_DES', 'ENTORNO_DES', 'SEGMENTO_MAESTRO_DESC', 'LID_UBICACION_TIENDA']
for col in cat_vars:
    le = LabelEncoder()
    df_bart[col] = le.fit_transform(df_bart[col])

# Preparar matrices
X = df_bart.drop(columns=['RATIO_CUMPLIMIENTO', 'TIENDA_ID'])
y = df_bart['RATIO_CUMPLIMIENTO'].values
tienda_ids = df_bart['TIENDA_ID'].values

# 2. Modelo BART mejorado

with pm.Model() as bart_model:
    # Configuración mejorada de BART
    μ = pb.BART(
        "μ",
        X=X.values,
        Y=y,
        m=100,
        alpha=0.95,
        beta=2.0,
        response="mix"
    )

    # Distribución t-Student (robusta a outliers)
    σ = pm.HalfStudentT("σ", nu=4, sigma=1)
    ν = pm.Gamma("ν", alpha=2, beta=0.1)
    y_obs = pm.StudentT("y_obs", mu=μ, sigma=σ, nu=ν, observed=y)

    # Muestreo MCMC
    trace = pm.sample(
        draws=1500,
        tune=1500,
        target_accept=0.98,
        chains=4,
        random_seed=42,
        idata_kwargs={"log_likelihood": True}
    )

    # Predicción posterior

    posterior_predictive = pm.sample_posterior_predictive(
        trace,
        var_names=["y_obs"],
        random_seed=42
    )

# 3. Procesamiento de resultados

idata = az.from_pymc3(trace=trace, posterior_predictive=posterior_predictive)

# Obtener predicciones
posterior_μ = idata.posterior["μ"]
y_pred_mean = posterior_μ.mean(dim=["chain", "draw"]).values
y_pred_std = posterior_μ.std(dim=["chain", "draw"]).values

# Calcular intervalos de credibilidad al 90%
y_lower = np.percentile(posterior_μ, 5, axis=(0, 1)).flatten()
y_upper = np.percentile(posterior_μ, 95, axis=(0, 1)).flatten()

# 4. Métricas de desempeño

mae = mean_absolute_error(y, y_pred_mean)
rmse = np.sqrt(mean_squared_error(y, y_pred_mean))
r2 = r2_score(y, y_pred_mean)
cobertura = ((y >= y_lower) & (y <= y_upper)).mean()

print(f"- MAE: {mae:.4f}")
print(f"- RMSE: {rmse:.4f}")
print(f"- R²: {r2:.4f}")
print(f"- Cobertura IC 90%: {cobertura:.1%}")

# 5. Visualizaciones
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (12, 6)

## 5.1 Gráfico de dispersión real vs predicho
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y, y=y_pred_mean, alpha=0.6, edgecolor='k')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', label='Línea perfecta')
plt.errorbar(x=y, y=y_pred_mean, yerr=y_pred_std,
             fmt='o', alpha=0.5, label='±1 desviación estándar')
plt.xlabel('Valor Real', fontsize=12)
plt.ylabel('Valor Predicho', fontsize=12)
plt.title('Dispersión: Valores Reales vs Predicciones BART', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

## 5.2 Gráfico de residuales
residuos = y - y_pred_mean
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_pred_mean, y=residuos, alpha=0.6)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel('Valor Predicho', fontsize=12)
plt.ylabel('Residual (Real - Predicho)', fontsize=12)
plt.title('Análisis de Residuales', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

## 5.3 Intervalos de credibilidad (20 tiendas aleatorias)
np.random.seed(42)
sample_idx = np.random.choice(len(y), 20, replace=False)
sample_tiendas = tienda_ids[sample_idx]

plt.figure(figsize=(14, 6))
plt.errorbar(
    x=sample_tiendas.astype(str),
    y=y_pred_mean[sample_idx],
    yerr=[y_pred_mean[sample_idx] - y_lower[sample_idx],
          y_upper[sample_idx] - y_pred_mean[sample_idx]],
    fmt='o', capsize=5, label='Predicción con IC 90%'
)
plt.scatter(
    sample_tiendas.astype(str),
    y[sample_idx],
    color='red',
    label='Valor Real'
)
plt.xticks(rotation=45)
plt.legend()
plt.title('Predicciones vs Valores Reales con Intervalos de Credibilidad (Muestra de 20 Tiendas)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

## 5.4 Distribución de los residuales
plt.figure(figsize=(10, 5))
sns.histplot(residuos, kde=True, bins=30)
plt.axvline(0, color='r', linestyle='--')
plt.xlabel('Residuales', fontsize=12)
plt.title('Distribución de los Residuales', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 6. Análisis de importancia de variables
print("\n🔍 Importancia de variables (aproximada):")
bart_var_imp = np.mean(trace['μ_tree_ensemble'][-1000:], axis=0)
var_imp_df = pd.DataFrame({
    'Variable': X.columns,
    'Importancia': bart_var_imp
}).sort_values('Importancia', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importancia', y='Variable', data=var_imp_df, palette='viridis')
plt.title('Importancia Relativa de Variables (BART)', fontsize=14)
plt.tight_layout()
plt.show()

print(" Análisis completo ejecutado correctamente")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import xgboost as xgb
from sklearn.utils import resample
from tqdm import tqdm
import joblib
import os

# 1. Preprocesamiento de datos
print("🔧 Preprocesando datos...")
variables_modelo = [
    'RATIO_CUMPLIMIENTO', 'TIENDA_ID',
    'dist_comp_directa', 'num_comp_directa',
    'dist_comp_indirecta', 'num_comp_indirecta',
    'NIVELSOCIOECONOMICO_DES', 'ENTORNO_DES',
    'MTS2VENTAS_NUM', 'PUERTASREFRIG_NUM',
    'LATITUD_NUM', 'LONGITUD_NUM',
    'SEGMENTO_MAESTRO_DESC', 'LID_UBICACION_TIENDA', 'META_VENTA'
]

# Cargar y limpiar datos
df_train = df_modelo[variables_modelo].dropna().drop_duplicates()

# Codificar variables categóricas
cat_vars = ['NIVELSOCIOECONOMICO_DES', 'ENTORNO_DES', 'SEGMENTO_MAESTRO_DESC', 'LID_UBICACION_TIENDA']
label_encoders = {}
for col in cat_vars:
    le = LabelEncoder()
    df_train[col] = le.fit_transform(df_train[col])
    label_encoders[col] = le

# Preparar matrices
X_train = df_train.drop(columns=['RATIO_CUMPLIMIENTO', 'TIENDA_ID'])
y_train = df_train['RATIO_CUMPLIMIENTO'].values
tienda_ids_train = df_train['TIENDA_ID'].values

# 2. Configuración y entrenamiento del modelo XGBoost
print("\n🎯 Entrenando modelo XGBoost con todos los datos disponibles...")
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'mae',
    'learning_rate': 0.05,
    'max_depth': 6,
    'min_child_weight': 3,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'n_estimators': 1000,
    'random_state': 42
}

model = xgb.XGBRegressor(**params)
model.fit(X_train, y_train)

# 3. Generación de intervalos de confianza mediante bootstrap
print("\n🔮 Generando intervalos de confianza con bootstrap...")
n_bootstrap = 100
bootstrap_preds = []

for _ in tqdm(range(n_bootstrap), desc="Bootstrap iterations"):
    X_resampled, y_resampled = resample(X_train, y_train, random_state=_)
    temp_model = xgb.XGBRegressor(**params)
    temp_model.fit(X_resampled, y_resampled, verbose=0)
    preds = temp_model.predict(X_train)
    bootstrap_preds.append(preds)

bootstrap_preds = np.array(bootstrap_preds)

# Calcular intervalos para los datos de entrenamiento
train_lower_bound = np.percentile(bootstrap_preds, 5, axis=0)  # IC 90%
train_upper_bound = np.percentile(bootstrap_preds, 95, axis=0)
train_pred_mean = np.mean(bootstrap_preds, axis=0)

# 4. Métricas de desempeño en entrenamiento
print("\n📈 Métricas de desempeño en datos de entrenamiento:")
mae_train = mean_absolute_error(y_train, train_pred_mean)
rmse_train = np.sqrt(mean_squared_error(y_train, train_pred_mean))
r2_train = r2_score(y_train, train_pred_mean)
cobertura_train = ((y_train >= train_lower_bound) & (y_train <= train_upper_bound)).mean()

print(f"- MAE: {mae_train:.4f}")
print(f"- RMSE: {rmse_train:.4f}")
print(f"- R²: {r2_train:.4f}")
print(f"- Cobertura IC 90%: {cobertura_train:.1%}")

# 5. Visualizaciones en datos de entrenamiento
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (12, 6)

## 5.1 Gráfico de dispersión real vs predicho (entrenamiento)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_train, y=train_pred_mean, alpha=0.6, edgecolor='k')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--', label='Línea perfecta')
plt.errorbar(x=y_train, y=train_pred_mean,
             yerr=[train_pred_mean - train_lower_bound, train_upper_bound - train_pred_mean],
             fmt='o', alpha=0.3, label='IC 90%')
plt.xlabel('Valor Real', fontsize=12)
plt.ylabel('Valor Predicho', fontsize=12)
plt.title('Dispersión: Valores Reales vs Predicciones (Datos de Entrenamiento)', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

## 5.2 Importancia de características
feature_importance = model.feature_importances_
sorted_idx = np.argsort(feature_importance)

plt.figure(figsize=(10, 8))
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), np.array(X_train.columns)[sorted_idx])
plt.xlabel('Importancia', fontsize=12)
plt.title('Importancia de Características (XGBoost)', fontsize=14)
plt.tight_layout()
plt.show()

# 6. Guardar modelo y componentes para predicción futura
print("\n💾 Guardando modelo y componentes para predicción...")
# Crear directorio si no existe
os.makedirs('modelos', exist_ok=True)

# Guardar componentes
joblib.dump({
    'model': model,
    'label_encoders': label_encoders,
    'feature_names': list(X_train.columns),
    'params': params
}, 'modelos/xgboost_model_full.pkl')

print("\n✅ Entrenamiento completo. Modelo guardado en: modelos/xgboost_model_full.pkl")
print("⚠️ Nota: Las métricas mostradas son sobre datos de entrenamiento (pueden estar optimistas)")
```
